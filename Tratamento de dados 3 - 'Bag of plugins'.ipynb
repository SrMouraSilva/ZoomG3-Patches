```python
import pandas as pd
import numpy as np
```

```python
pedalboards = pd.read_csv('data/pedalboard-plugin.csv')
pedalboards.columns = ['id', 'name', 0, 1, 2, 3, 4, 5]

pedalboards.head()
```

```python
plugins = pd.read_csv('data/plugin-category.csv', index_col='id')
plugins = plugins.sort_index()

plugins.head()
```

```python
data = pd.DataFrame([], index=pedalboards.index, columns=plugins.index)
data = data.fillna(0)

data.head()
```

```python
for index, pedalboard in pedalboards[[0, 1, 2, 3, 4, 5]].iterrows():
    for effect in pedalboard:
        if np.isnan(effect):
            continue
        data[effect][index] += 1

data.head()
```

```python
new_index = pd.MultiIndex.from_tuples(list(zip(pedalboards.index.tolist(), pedalboards['id'].tolist())), names=['index', 'id'])
data.index = new_index

data.head()
```

```python
data.to_csv('data/pedalboard-plugin-bag-of-words.csv')
```
