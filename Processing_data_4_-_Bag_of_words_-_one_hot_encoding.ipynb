```python
import pandas as pd
import numpy as np
```

```python
pedalboards = pd.read_csv('data/patches-filtered.csv')
pedalboards.columns = ['id', 'name', 0, 1, 2, 3, 4, 5]

pedalboards.head()
```

```python
plugins = pd.read_csv('data/plugins-categories.csv', index_col='id')
plugins = plugins.sort_index()

plugins.head()
```

# Simple bag of plugins

```python
data = pd.DataFrame([], index=pedalboards.index, columns=plugins.index)
data = data.fillna(0)

data.head()
```

```python
for index, pedalboard in pedalboards[[0, 1, 2, 3, 4, 5]].iterrows():
    for effect in pedalboard:
        if np.isnan(effect):
            continue
        data[effect][index] += 1

data.head()
```

```python
new_index = pd.MultiIndex.from_tuples(list(zip(pedalboards.index.tolist(), pedalboards['id'].tolist())), names=['index', 'id'])
data.index = new_index

data.head()
```

```python
data.to_csv('data/patches-bag-of-words.csv')
```

# One hot encoding

The position is considered.

```python
def extract_one_hot_encoding(position):
    data = pd.DataFrame([], index=pedalboards.index, columns=plugins.index)
    data = data.fillna(0)

    for index, pedalboard in pedalboards[[position]].iterrows():
        for effect in pedalboard:
            if np.isnan(effect):
                continue
            data[effect][index] += 1

    return data


bags = [extract_one_hot_encoding(i) for i in range(0, 6)]
```

```python
full_bag_of_plugins = pd.concat(bags, axis=1)

new_index = pd.MultiIndex.from_tuples(list(zip(pedalboards.index.tolist(), pedalboards['id'].tolist())), names=['index', 'id'])
full_bag_of_plugins.index = new_index

full_bag_of_plugins.to_csv('data/patches-one-hot-encoding.csv')
full_bag_of_plugins.head()
```
