# Tratamento de dados 1 - Plugins de áudio

## Categorias

```python
import json
import pandas as pd
```

```python
with open('data/plugins-metadata/plugins-parameters.json') as file:
    json_file = json.load(file)

data = []

for index, effect in enumerate(json_file['Effects']['Effect']):
    name = effect['-name']
    category = effect['-category']
    
    data.append((name, category))

data.append(('None', 'None'))

data = pd.DataFrame(data=data, columns=('name', 'category'))
data.head(10)
```

## Índices

```python
with open('data/plugins-metadata/plugins-index.json') as file:
    json_file = json.load(file)

indexes = []

for index, effect in enumerate(json_file['Modules']['UNIT1']['Type'][:-1]):
    index = int(effect['-EfxNoOnMachine'])
    name = effect['-name']
    
    indexes.append((index, name))

indexes = pd.DataFrame(data=indexes, columns=('id', 'pedal_name'))
indexes.head(10)
```

## Join tables

```python
plugins = pd.concat([data, indexes], axis=1)

plugins = plugins.set_index(plugins.id)
del plugins['id']
del plugins['pedal_name']

plugins.sort_index()
```

## Persistir

```python
import json

#with open(, 'w') as outfile:
plugins.to_csv('data/plugins_categories.csv')
```
